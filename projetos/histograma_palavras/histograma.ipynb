```python
def remover_tralha(texto):
    if isinstance(texto, str):
        texto = [texto]  # Converte a string em uma lista de strings
    
    resultado = []
    for string in texto:
        caracteres_alfanumericos = []
        for caractere in string:
            if caractere.isalnum():
                caracteres_alfanumericos.append(caractere)
        nova_string = ''.join(caracteres_alfanumericos)
        resultado.append(nova_string)
    
    if len(resultado) == 1:
        return resultado[0]  # Retorna a string diretamente se for apenas um elemento
    else:
        return resultado
```

```python
wordlist = list()
with open(filename) as f:
    lines = f.readlines()
    for line in lines:     
        line = line.lower()
        line = line.strip()
        words = line.split()
        words = remover_tralha(words)
        wordlist.extend(words)              

print('Linhas: ', len(lines))
print('Palavras: ', len(wordlist))

# wordlist.sort()
# print('Lista ordenada')
```
